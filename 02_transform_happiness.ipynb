```python
# import country&continent csv
country=os.path.join('00_input','continent_&_country_list.csv')
country_df=pd.read_csv(country)
```

```python
# drop duplicate and NaN country code and country
country_df=country_df.drop_duplicates(subset =["Three_Letter_Country_Code"], keep = 'last')
country_df=country_df.drop_duplicates(subset =["Country_x1"], keep = 'last')
country_df=country_df.dropna(axis=0)
country_df.head()
```

```python
# clean country name
for ind,row in country_df.iterrows():
    country_df.loc[ind,'Country_x1']=country_df.loc[ind,'Country_x1'].split(',')[0]
    country_df.loc[ind,'Country_x1']=country_df.loc[ind,'Country_x1'].split('(')[0]
country_df.head()
```

```python
# rename and rearrange columns
country_df_new=country_df[['Three_Letter_Country_Code','Country_x1','Continent_Name']]
country_df_new.columns=['countrycode','country','continent']
country_df_new.head()
```

```python
# save country csv for SQL database
country_df_new.to_csv('02_transform_happiness/country.csv',index=False)
```

```python
# replace countries with country code
for ind,row in happyall.iterrows():
    for ind2,row2 in country_df_new.iterrows():
        if row['country'].lower() in row2['country'].lower():
            happyall.loc[ind,'country']=country_df_new.loc[ind2,'countrycode']
            break
```

```python
# find countries without a country code from the main dataframe
miscountry=[]
for i in happyall['country']:
    if (len(i)>3) and i not in miscountry:
        miscountry.append(i)
print(miscountry)
```

```python
# invert match to capture countries that are named differently
# ie. previous round looking for happyall country string within continent_df country, 
# now looking for continent_df country string within happyall country
# takes care of situation such as happyall country of 'Congo (Brazzaville)' not in the continent_df country of 'Congo'
for ind,row in country_df_new.iterrows():
    for j in miscountry:
        if row['country'].lower() in j.lower():
            happyall['country']=happyall['country'].replace(j,country_df_new.loc[ind,'countrycode'])
```

```python
# find all missed countries again
miscountry=[]
for i in happyall['country']:
    if (len(i)>3) and i not in miscountry:
        miscountry.append(i)
print(miscountry)
```

```python
# last ditch attempt to get final few countries' alternate name by scrapping wiki
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from time import sleep
executable_path={'executable_path':'/usr/local/bin/chromedriver'}
browser=Browser('chrome',**executable_path,headless=False)
altname=[None]*len(miscountry)
for i in range(len(miscountry)):
    url='http://wikipedia.org'
    browser.visit(url)
    sleep(1)
    browser.find_by_id('searchInput').fill(f'{miscountry[i]}\n')
    html=browser.html
    soup=BeautifulSoup(html,'lxml')
    result=soup.find_all('div', class_='country-name')
    for r in result:
        altname[i]=r.text
browser.quit()
```
